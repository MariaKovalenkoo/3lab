#Бинарный классификатор
Модель 1

Создаем генераторы данных для каждого из этапов: train_dat, val_dat и test_dat. Они загружают изображения из указанных директорий,
Пиксели изображения должны быть масштабированы в диапазон от 0 до 1, делением на 255.

```python
df_path = '/content/gdrive/MyDrive/neironki/datasetBin/train'
df_path2 = '/content/gdrive/MyDrive/neironki/datasetBin/val'
df_path3 = '/content/gdrive/MyDrive/neironki/datasetBin/test'

train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)
train_dat =  train_datagen.flow_from_directory(df_path,batch_size=32, target_size = (256,256),class_mode='binary')
val_dat = val_datagen.flow_from_directory(df_path2, batch_size=32,target_size = (256,256),class_mode='binary')
test_dat = test_datagen.flow_from_directory(df_path3, batch_size=32, target_size = (256,256),class_mode='binary')
```

```python
from tensorflow.keras.layers import Input
from tensorflow.keras import layers
import tensorflow as tf

model = models.Sequential()
model.add(layers.Conv2D(32, (3,3), activation='relu',input_shape=(256, 256, 3)))
model.add(layers.Conv2D(32, (3,3), activation='relu'))
model.add(layers.MaxPooling2D(2, 2))
model.add(layers.Conv2D(64, (3,3), activation='relu'))
model.add(layers.MaxPooling2D(2, 2))
model.add(layers.Conv2D(128, (3,3), activation='relu'))
model.add(layers.MaxPooling2D(2, 2))
model.add(layers.Conv2D(128, (3,3), activation='relu'))
model.add(layers.MaxPooling2D(2, 2))
model.add(layers.Flatten())
model.add(layers.Dropout(0.5))

model.add(layers.Dense(1, activation='sigmoid'))
model.summary()
```

#Обучаем модель:

> Оптимизатор Adam

> функция потерь - binary_crossentropy  измеряет разницу между истинными метками и предсказаниями модели.





```python
model.compile(optimizer='adam', loss='binary_crossentropy',
              metrics='accuracy')
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping

early_stop = EarlyStopping(monitor='val_accuracy', mode='max', patience=10,
                           verbose=1)
checkpoint = ModelCheckpoint('/content/gdrive/MyDrive/neironki/best_model.h5',
                             monitor='val_accuracy', mode='max',
                             verbose=1, save_best_only=True)
callbacks_list = [early_stop, checkpoint]
```

```python
history1 = model.fit(train_dat, epochs=80,batch_size=32,validation_data=val_dat,callbacks=callbacks_list)
history_dict1 = history1.history
```

#Модель 2

rescale=1./255: Масштабирует каждый пиксель изображения, делая его значение от 0 до 1.

rotation_range=40: Изображения будут случайно поворачиваться на угол до 40 градусов.

width_shift_range=0.2 и height_shift_range=0.2: Изображения будут случайно сдвигаться по ширине и высоте на долю до 0.2 от общей ширины или высоты.

shear_range=0.2: Применяет случайные сдвиги силы сдвига.

zoom_range=0.2: Случайно увеличивает изображения до 20%.

horizontal_flip=True: Случайно переворачивает входы по горизонтали.

fill_mode='nearest': Стратегия, используемая для заполнения вновь появившихся пикселей, которые могут появиться после поворота или сдвига ширины/высоты.

```python
datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

train_aug = datagen.flow_from_directory(df_path3 ,target_size=(256, 256),
    batch_size=32,class_mode='binary')
val_aug = datagen.flow_from_directory(df_path3 ,target_size=(256, 256),
    batch_size=32,class_mode='binary')
```

Слой Input: Этот слой определяет форму входных данных (изображения размером 256x256 с 3 каналами цвета).

```python
from tensorflow.keras.models import Model
inputs = Input(shape=(256, 256, 3))

x = Conv2D(32, (3,3), activation='relu')(inputs)
x = Conv2D(64, (3,3), activation='relu')(x)
x = MaxPooling2D(2, 2)(x)
x = Flatten()(x)
x = Dense(1, activation='sigmoid')(x)
model2 = Model(inputs=inputs, outputs=x)

model2.summary()
```

Обучаем модель

функция потерь - binary_crossentropy часто используется в задачах бинарной классификации, измеряет разницу между истинными метками и предсказаниями модели.

```python
model2.compile(optimizer='adam', loss='binary_crossentropy',
              metrics='accuracy')
```
